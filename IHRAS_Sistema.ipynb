## 🔧 Instalação e Verificação de Dependências

Esta célula instala automaticamente todas as dependências necessárias e verifica se o **pytorch-grad-cam** está funcionando corretamente.

```python
# 🔧 IHRAS Medical Intelligence - Instalação e Verificação
# Prof. Dr. André Luiz Marques Serrano (Instituto de Ciências Matemáticas e Computacionais - USP)

import subprocess
import sys
import importlib
import warnings
warnings.filterwarnings('ignore')

print("🏥 IHRAS Medical Intelligence v9.0")
print("🔧 Iniciando instalação e verificação de dependências...\n")

def install_and_verify_package(package_name, import_name=None, test_function=None):
    """Instala e verifica se um pacote está funcionando corretamente"""
    if import_name is None:
        import_name = package_name.replace('-', '_')

    try:
        # Tentar importar primeiro
        module = importlib.import_module(import_name)
        print(f"✅ {package_name} já está instalado")

        # Executar teste se fornecido
        if test_function:
            test_result = test_function(module)
            if test_result:
                print(f"✅ {package_name} verificado e funcionando")
            else:
                print(f"⚠️ {package_name} instalado mas com problemas")

        return True

    except ImportError:
        print(f"📦 Instalando {package_name}...")
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", package_name, "--quiet"])

            # Verificar instalação
            module = importlib.import_module(import_name)
            print(f"✅ {package_name} instalado com sucesso")

            # Executar teste se fornecido
            if test_function:
                test_result = test_function(module)
                if test_result:
                    print(f"✅ {package_name} verificado e funcionando")
                else:
                    print(f"⚠️ {package_name} instalado mas com problemas")

            return True

        except Exception as e:
            print(f"❌ Erro ao instalar {package_name}: {e}")
            return False

def test_pytorch_grad_cam(module):
    """Testa se o pytorch-grad-cam está funcionando corretamente"""
    try:
        from pytorch_grad_cam import GradCAM
        from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
        from pytorch_grad_cam.utils.image import show_cam_on_image

        # Teste básico de importação
        print("  🔍 Testando importações do Grad-CAM...")

        # Verificar se as classes principais existem
        assert hasattr(GradCAM, '__init__'), "GradCAM class não encontrada"
        assert callable(ClassifierOutputTarget), "ClassifierOutputTarget não é callable"
        assert callable(show_cam_on_image), "show_cam_on_image não é callable"

        print("  ✅ Todas as importações do Grad-CAM funcionando")
        return True

    except Exception as e:
        print(f"  ❌ Erro no teste do Grad-CAM: {e}")
        return False

def test_torch(module):
    """Testa se o PyTorch está funcionando corretamente"""
    try:
        import torch
        print(f"  🔍 PyTorch versão: {torch.__version__}")
        print(f"  🔍 CUDA disponível: {'Sim' if torch.cuda.is_available() else 'Não'}")
        if torch.cuda.is_available():
            print(f"  🔍 GPU: {torch.cuda.get_device_name(0)}")

        # Teste básico de tensor
        x = torch.randn(2, 3)
        y = x + 1
        assert y.shape == (2, 3), "Operação básica de tensor falhou"

        return True
    except Exception as e:
        print(f"  ❌ Erro no teste do PyTorch: {e}")
        return False

# Lista de dependências com testes
dependencies = [
    ("torch", "torch", test_torch),
    ("torchvision", "torchvision", None),
    ("grad-cam", "pytorch_grad_cam", test_pytorch_grad_cam),
    ("pillow", "PIL", None),
    ("opencv-python", "cv2", None),
    ("matplotlib", "matplotlib", None),
    ("seaborn", "seaborn", None),
    ("numpy", "numpy", None),
    ("pandas", "pandas", None),
    ("scikit-learn", "sklearn", None),
    ("ipywidgets", "ipywidgets", None),
    ("markdown", "markdown", None),
    ("fpdf2", "fpdf", None)
]

# Instalar e verificar cada dependência
success_count = 0
total_count = len(dependencies)

for package_name, import_name, test_func in dependencies:
    if install_and_verify_package(package_name, import_name, test_func):
        success_count += 1
    print()  # Linha em branco

print(f"📊 Resumo da Instalação:")
print(f"✅ Sucesso: {success_count}/{total_count} pacotes")

if success_count == total_count:
    print("🎉 Todas as dependências instaladas e verificadas com sucesso!")
    print("🏥 Sistema IHRAS pronto para uso!")
else:
    print(f"⚠️ {total_count - success_count} pacotes com problemas")
    print("🔧 Verifique os erros acima e tente novamente")

print("\n" + "="*60)
```

## 🤖 Sistema Principal IHRAS

Implementação da classe principal do sistema de análise radiológica.

```python
# 🤖 Classe Principal do Sistema IHRAS
class IHRASMedicalSystem:
    """
    Sistema IHRAS - Intelligent Humanized Radiology Analysis System
    Prof. Dr. André Luiz Marques Serrano (Instituto de Ciências Matemáticas e Computacionais - USP)
    """

    def __init__(self):
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.model = None
        self.transform = None
        self.pathologies = list(PATHOLOGIES_DATA.keys())
        self.setup_model()
        self.setup_transforms()

        print(f"🏥 IHRAS Medical Intelligence v9.0 inicializado")
        print(f"🔧 Dispositivo: {self.device}")
        print(f"📊 Patologias detectáveis: {len(self.pathologies)}")

    def setup_model(self):
        """Configura o modelo DenseNet121 para análise radiológica"""
        try:
            print("🔧 Configurando modelo DenseNet121...")

            # Carregar DenseNet121 pré-treinado
            self.model = models.densenet121(pretrained=True)

            # Modificar a camada final para 14 patologias
            num_features = self.model.classifier.in_features
            self.model.classifier = nn.Linear(num_features, len(self.pathologies))

            # Mover para dispositivo
            self.model = self.model.to(self.device)
            self.model.eval()

            print("✅ Modelo DenseNet121 configurado com sucesso")
            print(f"📊 Parâmetros: {sum(p.numel() for p in self.model.parameters()):,}")

        except Exception as e:
            print(f"❌ Erro ao configurar modelo: {e}")
            raise

    def setup_transforms(self):
        """Configura as transformações de imagem"""
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]
            )
        ])
        print("✅ Transformações de imagem configuradas")

    def preprocess_image(self, image):
        """Pré-processa a imagem para análise"""
        if isinstance(image, str):
            image = Image.open(image)

        # Converter para RGB se necessário
        if image.mode != 'RGB':
            image = image.convert('RGB')

        # Aplicar transformações
        tensor = self.transform(image).unsqueeze(0).to(self.device)

        return tensor, image

    def analyze_pathologies(self, image_tensor):
        """Analisa patologias na imagem"""
        with torch.no_grad():
            outputs = self.model(image_tensor)
            probabilities = torch.sigmoid(outputs).cpu().numpy()[0]

        # Criar dicionário de resultados
        results = {}
        for i, pathology in enumerate(self.pathologies):
            confidence = float(probabilities[i])
            threshold = PATHOLOGIES_DATA[pathology]["threshold"]

            results[pathology] = {
                "confidence": confidence,
                "detected": confidence >= threshold,
                "threshold": threshold,
                "urgencia": PATHOLOGIES_DATA[pathology]["urgencia"],
                "nome_pt": PATHOLOGIES_DATA[pathology]["nome_pt"],
                "grupo": PATHOLOGIES_DATA[pathology]["grupo"],
                "descricao": PATHOLOGIES_DATA[pathology]["descricao"]
            }

        return results

    def generate_gradcam(self, image_tensor, original_image, pathology_results):
        """Gera mapas de calor Grad-CAM para patologias detectadas"""
        try:
            print("🔍 Gerando mapas Grad-CAM...")

            # Identificar patologias detectadas
            detected_pathologies = [
                path for path, result in pathology_results.items()
                if result["detected"]
            ]

            if not detected_pathologies:
                print("ℹ️ Nenhuma patologia detectada para Grad-CAM")
                return None

            # Configurar Grad-CAM
            target_layer = [self.model.features.denseblock4.denselayer16.conv2]

            # Criar instância do GradCAM (sem parâmetro use_cuda)
            cam = GradCAM(model=self.model, target_layers=target_layer)

            # Gerar heatmap para a patologia com maior confiança
            best_pathology = max(detected_pathologies,
                               key=lambda x: pathology_results[x]["confidence"])

            pathology_idx = self.pathologies.index(best_pathology)
            targets = [ClassifierOutputTarget(pathology_idx)]

            # Gerar CAM
            grayscale_cam = cam(input_tensor=image_tensor, targets=targets)
            grayscale_cam = grayscale_cam[0, :]

            # Converter imagem original para array numpy
            original_array = np.array(original_image.resize((224, 224))) / 255.0

            # Sobrepor CAM na imagem
            visualization = show_cam_on_image(original_array, grayscale_cam, use_rgb=True)

            print(f"✅ Grad-CAM gerado para: {PATHOLOGIES_DATA[best_pathology]['nome_pt']}")

            return {
                "heatmap": visualization,
                "pathology": best_pathology,
                "pathology_pt": PATHOLOGIES_DATA[best_pathology]['nome_pt'],
                "confidence": pathology_results[best_pathology]["confidence"]
            }

        except Exception as e:
            print(f"⚠️ Erro ao gerar Grad-CAM: {e}")
            return None

    def complete_analysis(self, image, filename="", patient_info=""):
        """Executa análise completa da radiografia"""
        try:
            print("🔍 Iniciando análise IHRAS...")

            # Pré-processar imagem
            image_tensor, original_image = self.preprocess_image(image)
            print("✅ Imagem pré-processada")

            # Analisar patologias
            pathology_results = self.analyze_pathologies(image_tensor)
            print("✅ Análise de patologias concluída")

            # Gerar Grad-CAM
            gradcam_result = self.generate_gradcam(image_tensor, original_image, pathology_results)

            # Determinar nível de alerta
            alert_level = self.determine_alert_level(pathology_results)
            print(f"🚨 Nível de alerta: {alert_level['level']}")

            # Compilar resultados
            analysis_result = {
                "timestamp": datetime.now().isoformat(),
                "filename": filename,
                "patient_info": patient_info,
                "pathologies": pathology_results,
                "gradcam": gradcam_result,
                "alert_level": alert_level,
                "system_info": {
                    "version": "9.0",
                    "model": "DenseNet121",
                    "device": str(self.device)
                }
            }

            print("✅ Análise IHRAS concluída com sucesso!")
            return analysis_result

        except Exception as e:
            print(f"❌ Erro na análise: {e}")
            return None

    def determine_alert_level(self, pathology_results):
        """Determina o nível de alerta baseado nas patologias detectadas"""
        detected = [p for p, r in pathology_results.items() if r["detected"]]

        if not detected:
            return {"level": "Normal", "color": COLORS['success'], "icon": "✅"}

        # Verificar urgências
        urgencies = [pathology_results[p]["urgencia"] for p in detected]

        if "Emergência" in urgencies:
            return {"level": "CRÍTICO", "color": COLORS['danger'], "icon": "🚨"}
        elif "Urgente" in urgencies:
            return {"level": "URGENTE", "color": COLORS['danger'], "icon": "⚠️"}
        elif "Alta" in urgencies:
            return {"level": "ALTO", "color": COLORS['warning'], "icon": "🔶"}
        elif "Moderada" in urgencies:
            return {"level": "MODERADO", "color": COLORS['warning'], "icon": "🔸"}
        else:
            return {"level": "BAIXO", "color": COLORS['success'], "icon": "🔹"}

print("✅ Classe IHRASMedicalSystem definida com sucesso!")
```

Sistema de geração de laudos médicos em formato Markdown e PDF.

```python
# 📄 Gerador de Laudos IHRAS
class IHRASReportGenerator:
    """Gerador de laudos médicos estruturados para o sistema IHRAS"""

    def __init__(self):
        self.colors = COLORS

    def generate_markdown_report(self, analysis_result):
        """Gera laudo médico em formato Markdown"""

        timestamp = datetime.now().strftime("%d/%m/%Y às %H:%M:%S")
        analysis_id = str(uuid.uuid4())[:8]

        # Cabeçalho do laudo
        report = f"""# 🏥 LAUDO RADIOLÓGICO COMPUTADORIZADO

**Sistema:** IHRAS Medical Intelligence v9.0
**Pesquisador Responsável:** Prof. Dr. André Luiz Marques Serrano
**Instituição:** Instituto de Ciências Matemáticas e Computacionais - USP
**Data/Hora:** {timestamp}
**ID da Análise:** {analysis_id}

---

## 📋 DADOS DO PACIENTE

**Paciente:** {analysis_result.get('patient_info', 'Não informado')}
**Arquivo:** {analysis_result.get('filename', 'Não informado')}
**Tipo de Exame:** Radiografia de Tórax PA
**Sistema Utilizado:** IHRAS Medical Intelligence v9.0

---

## 🚨 NÍVEL DE ALERTA CLÍNICO

**Status:** {analysis_result['alert_level']['icon']} **{analysis_result['alert_level']['level']}**

"""

        # Resumo clínico baseado no nível de alerta
        if analysis_result['alert_level']['level'] == "CRÍTICO":
            report += "⚠️ **ATENÇÃO:** Achados sugestivos de condição que requer avaliação médica IMEDIATA.\n\n"
        elif analysis_result['alert_level']['level'] == "URGENTE":
            report += "⚠️ **IMPORTANTE:** Achados que requerem avaliação médica prioritária.\n\n"
        elif analysis_result['alert_level']['level'] == "ALTO":
            report += "📋 **RELEVANTE:** Achados que requerem correlação clínica e seguimento.\n\n"
        elif analysis_result['alert_level']['level'] == "MODERADO":
            report += "📋 **OBSERVAÇÃO:** Achados que podem requerer acompanhamento clínico.\n\n"
        elif analysis_result['alert_level']['level'] == "BAIXO":
            report += "📋 **ACHADOS MENORES:** Alterações de menor significado clínico.\n\n"
        else:
            report += "✅ **NORMAL:** Nenhuma alteração significativa detectada pelo sistema.\n\n"

        # Achados radiológicos
        report += "## 🔍 ACHADOS RADIOLÓGICOS\n\n"

        detected_pathologies = [
            (path, result) for path, result in analysis_result['pathologies'].items()
            if result['detected']
        ]

        if detected_pathologies:
            report += "### Patologias Detectadas:\n\n"
            for pathology, result in detected_pathologies:
                urgency_icon = self.get_urgency_icon(result['urgencia'])
                report += f"- **{result['nome_pt']}** {urgency_icon}\n"
                report += f"  - Confiança: {result['confidence']:.1%}\n"
                report += f"  - Grupo: {result['grupo']}\n"
                report += f"  - Descrição: {result['descricao']}\n\n"
        else:
            report += "✅ **Nenhuma patologia significativa detectada pelo sistema.**\n\n"

        # Tabela de análise de confiança
        report += "## 📊 ANÁLISE DE CONFIANÇA POR PATOLOGIA\n\n"
        report += "| Patologia | Confiança | Status | Urgência |\n"
        report += "|:----------|:---------:|:------:|:--------:|\n"

        for pathology, result in analysis_result['pathologies'].items():
            status = "✅ Detectada" if result['detected'] else "❌ Não detectada"
            urgency_icon = self.get_urgency_icon(result['urgencia'])
            report += f"| {result['nome_pt']} | {result['confidence']:.1%} | {status} | {urgency_icon} |\n"

        # Recomendações clínicas
        report += "\n## 💡 RECOMENDAÇÕES CLÍNICAS\n\n"

        if detected_pathologies:
            high_urgency = [p for p, r in detected_pathologies if r['urgencia'] in ['Emergência', 'Urgente']]
            if high_urgency:
                report += "🚨 **URGENTE:** Recomenda-se avaliação médica imediata devido aos achados detectados.\n\n"

            report += "📋 **Correlação Clínica:** Os achados radiológicos devem ser correlacionados com:\n"
            report += "- História clínica detalhada\n"
            report += "- Exame físico completo\n"
            report += "- Exames laboratoriais quando indicados\n"
            report += "- Outros exames de imagem se necessário\n\n"

            report += "🔄 **Seguimento:** Considerar reavaliação radiológica conforme evolução clínica.\n\n"
        else:
            report += "✅ **Exame Normal:** Nenhuma alteração significativa detectada.\n"
            report += "📋 **Recomendação:** Manter seguimento clínico de rotina conforme indicação médica.\n\n"

        # Informações técnicas
        report += "## 🔧 INFORMAÇÕES TÉCNICAS\n\n"
        report += f"- **Modelo de IA:** {analysis_result['system_info']['model']}\n"
        report += f"- **Dispositivo:** {analysis_result['system_info']['device']}\n"
        report += f"- **Versão do Sistema:** {analysis_result['system_info']['version']}\n"
        report += f"- **Timestamp:** {analysis_result['timestamp']}\n\n"

        # Referências científicas
        report += "## 📚 REFERÊNCIAS\n\n"
        report += "[1] Kufel, J., et al. (2023). Multi-Label Classification of Chest X-ray Abnormalities Using Transfer Learning Techniques. *Journal of Personalized Medicine*, 13(10), 1426. Disponível em: [https://pmc.ncbi.nlm.nih.gov/articles/PMC10607847/](https://pmc.ncbi.nlm.nih.gov/articles/PMC10607847/)\n\n"
        report += "[2] Smithuis, R. (2014). Chest X-Ray - Lung disease: Four-Pattern Approach. *The Radiology Assistant*. Disponível em: [https://radiologyassistant.nl/chest/chest-x-ray/lung-disease](https://radiologyassistant.nl/chest/chest-x-ray/lung-disease)\n\n"

        # Aviso médico
        report += "## ⚠️ AVISO MÉDICO\n\n"
        report += "**IMPORTANTE:** Este laudo foi gerado por sistema de inteligência artificial e deve ser utilizado apenas como ferramenta de apoio ao diagnóstico radiológico. A interpretação final e a decisão clínica são sempre de responsabilidade do médico assistente. Recomenda-se correlação clínica obrigatória e validação por radiologista qualificado.\n\n"

        report += "---\n"
        report += "*© 2025 - IHRAS Medical Intelligence - Prof. Dr. André Luiz Marques Serrano (Instituto de Ciências Matemáticas e Computacionais - USP)*"

        return report

    def get_urgency_icon(self, urgency):
        """Retorna ícone baseado no nível de urgência"""
        icons = {
            "Emergência": "🔴",
            "Urgente": "🔴",
            "Alta": "🟠",
            "Moderada": "🟡",
            "Baixa": "🟢"
        }
        return icons.get(urgency, "⚪")

    def save_report(self, report_content, filename_prefix="IHRAS_Laudo"):
        """Salva o relatório em arquivo"""
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"{filename_prefix}_{timestamp}.md"

        with open(filename, 'w', encoding='utf-8') as f:
            f.write(report_content)

        return filename

print("✅ Classe IHRASReportGenerator definida com sucesso!")
```

## 🔍 Função de Análise Completa

Implementação da função que executa a análise completa e exibe os resultados.

```python
# 🔍 Função de Análise Completa

def perform_analysis():
    """Executa análise completa da radiografia"""

    with output_area:
        clear_output(wait=True)

        # Verificar se há arquivo carregado
        if not file_upload.value:
            display(HTML('<div style="color: red; font-weight: bold;">❌ Por favor, faça upload de uma radiografia!</div>'))
            return

        # Obter informações do paciente
        patient_info = f"{patient_name.value}, {patient_age.value} anos, {patient_sex.value}"
        if patient_id.value:
            patient_info += f", ID: {patient_id.value}"

        try:
            # Exibir progresso
            display(HTML('''
            <div class="ihras-card">
                <h3>🔄 Processando Análise IHRAS...</h3>
                <div style="background: #f3f2f1; border-radius: 4px; padding: 10px; margin: 10px 0;">
                    <div style="background: #0078d4; height: 20px; border-radius: 4px; width: 0%; transition: width 2s;" id="progress"></div>
                </div>
            </div>
            '''))

            # Carregar imagem
            filename = list(file_upload.value.keys())[0]
            image_data = file_upload.value[filename]['content']
            image = Image.open(io.BytesIO(image_data))

            # Inicializar sistema IHRAS
            print("🤖 Inicializando sistema IHRAS...")
            ihras_system = IHRASMedicalSystem()

            # Executar análise
            print("🔍 Executando análise...")
            analysis_result = ihras_system.complete_analysis(image, filename, patient_info)

            if not analysis_result:
                display(HTML('<div style="color: red;">❌ Erro na análise. Tente novamente.</div>'))
                return

            # Limpar output e exibir resultados
            clear_output(wait=True)

            # Cabeçalho dos resultados
            alert_level = analysis_result['alert_level']
            alert_class = f"alert-{alert_level['level'].lower()}"

            display(HTML(f'''
            <div class="ihras-card {alert_class}">
                <h2>{alert_level['icon']} RESULTADO DA ANÁLISE IHRAS</h2>
                <h3>Nível de Alerta: {alert_level['level']}</h3>
                <p><strong>Paciente:</strong> {patient_info}</p>
                <p><strong>Arquivo:</strong> {filename}</p>
                <p><strong>Data/Hora:</strong> {datetime.now().strftime("%d/%m/%Y às %H:%M:%S")}</p>
            </div>
            '''))

            # Exibir imagem original e Grad-CAM
            fig, axes = plt.subplots(1, 2, figsize=(15, 6))

            # Imagem original
            axes[0].imshow(image, cmap='gray')
            axes[0].set_title('Radiografia Original', fontsize=14, fontweight='bold')
            axes[0].axis('off')

            # Grad-CAM
            if analysis_result['gradcam']:
                gradcam_data = analysis_result['gradcam']
                axes[1].imshow(gradcam_data['heatmap'])
                axes[1].set_title(f'Grad-CAM: {gradcam_data["pathology_pt"]}\n(Confiança: {gradcam_data["confidence"]:.1%})',
                                fontsize=14, fontweight='bold')
            else:
                axes[1].imshow(image, cmap='gray')
                axes[1].set_title('Grad-CAM: Nenhuma patologia detectada', fontsize=14, fontweight='bold')

            axes[1].axis('off')

            plt.tight_layout()
            plt.show()

            # Tabela de resultados
            detected_pathologies = [
                (path, result) for path, result in analysis_result['pathologies'].items()
                if result['detected']
            ]

            if detected_pathologies:
                display(HTML('<div class="ihras-card"><h3>🔍 Patologias Detectadas</h3></div>'))

                for pathology, result in detected_pathologies:
                    urgency_color = {
                        'Emergência': '#d13438',
                        'Urgente': '#d13438',
                        'Alta': '#ffb900',
                        'Moderada': '#ffb900',
                        'Baixa': '#107c10'
                    }.get(result['urgencia'], '#323130')

                    display(HTML(f'''
                    <div style="border-left: 4px solid {urgency_color}; padding: 10px; margin: 10px 0; background: #f9f9f9;">
                        <h4 style="margin: 0; color: {urgency_color};">{result['nome_pt']}</h4>
                        <p><strong>Confiança:</strong> {result['confidence']:.1%}</p>
                        <p><strong>Urgência:</strong> {result['urgencia']}</p>
                        <p><strong>Grupo:</strong> {result['grupo']}</p>
                        <p><strong>Descrição:</strong> {result['descricao']}</p>
                    </div>
                    '''))
            else:
                display(HTML('''
                <div class="ihras-card" style="background: #107c10; color: white;">
                    <h3>✅ Exame Normal</h3>
                    <p>Nenhuma patologia significativa detectada pelo sistema IHRAS.</p>
                </div>
                '''))

            # Gerar e exibir laudo
            print("📄 Gerando laudo médico...")
            report_generator = IHRASReportGenerator()
            markdown_report = report_generator.generate_markdown_report(analysis_result)

            # Salvar laudo
            report_filename = report_generator.save_report(markdown_report)

            display(HTML(f'''
            <div class="ihras-card">
                <h3>📄 Laudo Médico Gerado</h3>
                <p>✅ Laudo salvo como: <strong>{report_filename}</strong></p>
                <p>📋 O laudo completo está disponível para download.</p>
            </div>
            '''))

            # Exibir preview do laudo
            display(Markdown(markdown_report))

        except Exception as e:
            display(HTML(f'<div style="color: red; font-weight: bold;">❌ Erro na análise: {str(e)}</div>'))
            print(f"Erro detalhado: {e}")

# Conectar função ao botão
analyze_button.on_click(lambda b: perform_analysis())

print("✅ Função de análise configurada!")
print("🎯 Sistema IHRAS pronto para uso!")
```
